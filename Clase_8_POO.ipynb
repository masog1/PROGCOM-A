```python
def inventario():
  flores=["tulipan","lirio","rosa","dahlia","margarita","girasol","clavel"]
  seleccion=random.sample(flores,3)

  inventario={
      flor:{
          "precio":random.randint(2000,6000),
          "cantidad":random.randint(10,50),
          "vendidas":0
      } for flor in seleccion
      }
  return inventario
```
