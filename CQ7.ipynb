```python
# 1 for → sirve para recorrer una secuencia o rango.
for i in range(3):
    print(i)

# 2 def → define una función.
def saludo():
    print("Hola")
saludo()

# 3 return → devuelve un valor desde una función.
def suma(a, b):
    return a + b
print(suma(2, 3))

# 4 while → repite un bloque mientras la condición sea verdadera.
c = 0
while c < 3:
    print(c)
    c += 1

# 5 True → valor booleano verdadero.
x = True
print(x)

# 6 False → valor booleano falso.
x = False
print(x)

# 7 None → representa ausencia de valor.
x = None
if x is None:
    print("Sin valor")

# 8 and → operador lógico “y”.
print(True and False)

# 9 or → operador lógico “o”.
print(True or False)

# 10 not → operador lógico de negación.
print(not True)

# 11 if → condición básica.
if 5 > 3:
    print("Sí")

# 12 elif → condición alternativa en if.
x = 10
if x < 5:
    print("Pequeño")
elif x == 10:
    print("Diez")

# 13 else → bloque final si no se cumplen condiciones.
x = 2
if x > 5:
    print("Grande")
else:
    print("Pequeño")

# 14 break → detiene un bucle.
for i in range(5):
    if i == 3:
        break
    print(i)

# 15 continue → salta a la siguiente iteración.
for i in range(5):
    if i == 2:
        continue
    print(i)

# 16 pass → instrucción vacía.
def funcion():
    pass

# 17 try → bloque para probar código que puede fallar.
try:
    print(10/0)
except:
    print("Error")

# 18 except → maneja la excepción ocurrida.
try:
    print(10/0)
except ZeroDivisionError:
    print("No se puede dividir por cero")

# 19 finally → bloque que siempre se ejecuta.
try:
    x = 1/0
except:
    print("Error")
finally:
    print("Fin del programa")

# 20 raise → lanza un error manualmente.
raise ValueError("Error provocado")

# 21 assert → asegura que una condición sea verdadera.
x = 5
assert x > 0, "Debe ser positivo"

# 22 lambda → define una función anónima.
cuadrado = lambda x: x**2
print(cuadrado(4))

# 23 global → accede/modifica variables globales.
x = 10
def f():
    global x
    x = 20
f()
print(x)

# 24 nonlocal → accede a variables de una función externa.
def externo():
    x = 5
    def interno():
        nonlocal x
        x = 10
    interno()
    print(x)
externo()

# 25 class → define una clase.
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre
p = Persona("Ana")
print(p.nombre)

# 26 is → compara identidad de objetos.
a = [1, 2]
b = a
print(a is b)

# 27 import → importa un módulo completo.
import math
print(math.sqrt(16))

# 28 from → importa solo partes de un módulo.
from math import pi
print(pi)

# 29 as → crea un alias para un módulo.
import math as m
print(m.sqrt(9))

# 30 in → comprueba pertenencia.
print(2 in [1, 2, 3])

# 31 with → maneja recursos automáticamente.
with open("archivo.txt", "w") as f:
    f.write("Hola")

# 32 yield → convierte función en generador.
def contar():
    for i in range(3):
        yield i
for n in contar():
    print(n)

# 33 del → elimina una variable u objeto.
x = 10
del x

# 34 print → muestra datos en pantalla.
print("Hola mundo")

# 35 input → permite ingresar datos desde teclado.
# nombre = input("Tu nombre: ")
# print("Hola", nombre)

# 36 len → devuelve la longitud de un objeto.
print(len([1, 2, 3]))

# 37 open → abre un archivo.
f = open("archivo.txt", "w")
f.write("Probando")
f.close()

# 38 range → genera una secuencia de números.
for i in range(5):
    print(i)

# 39 enumerate → obtiene índice y valor al recorrer.
for i, v in enumerate(["a", "b"]):
    print(i, v)

# 40 zip → combina elementos de varias listas.
a = [1, 2]
b = ["uno", "dos"]
print(list(zip(a, b)))

# 41 map → aplica función a cada elemento.
nums = [1, 2, 3]
print(list(map(lambda x: x*2, nums)))

# 42 filter → filtra elementos según condición.
nums = [1, 2, 3, 4]
print(list(filter(lambda x: x%2==0, nums)))

# 43 sorted → devuelve lista ordenada.
print(sorted([3, 1, 2]))

# 44 reversed → invierte una secuencia.
print(list(reversed([1, 2, 3])))

# 45 super → accede a métodos de la clase padre.
class A:
    def saludo(self):
        print("Hola desde A")
class B(A):
    def saludo(self):
        super().saludo()
        print("Hola desde B")
B().saludo()

# 46 __init__ → constructor de una clase.
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre
p = Persona("Ana")
print(p.nombre)

# 47 dir → lista atributos y métodos disponibles.
print(dir([]))

# 48 help → muestra ayuda de objetos/módulos.
help(print)

# 49 id → devuelve la dirección en memoria de un objeto.
x = 5
print(id(x))

# 50 type → muestra el tipo de dato de un objeto.
print(type(3.14))
```
