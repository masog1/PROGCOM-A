```python
# 1 for → sirve para recorrer una secuencia o rango.
for i in range(3):
    print(i)

# 2 def → define una función.
def saludo():
    print("Hola")
saludo()

# 3 return → devuelve un valor desde una función.
def suma(a, b):
    return a + b
print(suma(2, 3))

# 4 while → repite un bloque mientras la condición sea verdadera.
c = 0
while c < 3:
    print(c)
    c += 1

# 5 True → valor booleano verdadero.
x = True
print(x)

# 6 False → valor booleano falso.
x = False
print(x)

# 7 None → representa ausencia de valor.
x = None
if x is None:
    print("Sin valor")

# 8 and → operador lógico “y”.
print(True and False)

# 9 or → operador lógico “o”.
print(True or False)

# 10 not → operador lógico de negación.
print(not True)

# 11 if → condición básica.
if 5 > 3:
    print("Sí")

# 12 elif → condición alternativa en if.
x = 10
if x < 5:
    print("Pequeño")
elif x == 10:
    print("Diez")

# 13 else → bloque final si no se cumplen condiciones.
x = 2
if x > 5:
    print("Grande")
else:
    print("Pequeño")

# 14 break → detiene un bucle.
for i in range(5):
    if i == 3:
        break
    print(i)

# 15 continue → salta a la siguiente iteración.
for i in range(5):
    if i == 2:
        continue
    print(i)

# 16 pass → instrucción vacía.
def funcion():
    pass
funcion()

# 17 try → bloque para probar código que puede fallar.
try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("División por cero detectada")
else:
    print("Resultado seguro:", resultado)
finally:
    print("Bloque finally ejecutado")

# 18 except → maneja la excepción ocurrida.
try:
    val = int("42")
except ValueError:
    print("No es un entero")
else:
    print("Conversión correcta:", val)

# 19 finally → bloque que siempre se ejecuta.
try:
    a = 1 + 1
except Exception:
    print("No debería ocurrir")
finally:
    print("Siempre se ejecuta (ejemplo seguro)")

# 20 raise → lanza un error manualmente (ejemplo controlado: no se ejecuta)
def maybe_raise(do_raise=False):
    if do_raise:
        raise ValueError("Error provocado (no activado)")
    return "OK"
print(maybe_raise(False))

# 21 assert → asegura que una condición sea verdadera.
x = 5
assert x > 0, "Debe ser positivo"
print("Assert comprobado correctamente")

# 22 lambda → define una función anónima.
cuadrado = lambda x: x**2
print(cuadrado(4))

# 23 global → accede/modifica variables globales.
x = 10
def f():
    global x
    x = 20
f()
print(x)

# 24 nonlocal → accede a variables de una función externa.
def externo():
    x = 5
    def interno():
        nonlocal x
        x = 10
    interno()
    print(x)
externo()

# 25 class → define una clase.
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre
p = Persona("Ana")
print(p.nombre)

# 26 is → compara identidad de objetos.
a = [1, 2]
b = a
print(a is b)

# 27 import → importa un módulo completo.
import math
print(math.sqrt(16))

# 28 from → importa solo partes de un módulo.
from math import pi
print(pi)

# 29 as → crea un alias para un módulo.
import math as m
print(m.sqrt(9))

# 30 in → comprueba pertenencia.
print(2 in [1, 2, 3])

# 31 with → maneja recursos automáticamente.
with open("archivo.txt", "w", encoding="utf-8") as f:
    f.write("Hola\n")

# 32 yield → convierte función en generador.
def contar():
    for i in range(3):
        yield i
for n in contar():
    print(n)

# 33 del → elimina una variable u objeto.
x = 10
del x
# (ya no existe x; no intentamos usarla después)

# 34 print → muestra datos en pantalla.
print("Hola mundo")

# 35 input → permite ingresar datos desde teclado.
# (Se muestra comentado para que el script no bloquee en entornos no interactivos)
# nombre = input("Tu nombre: ")
# print("Hola", nombre)

# 36 len → devuelve la longitud de un objeto.
print(len([1, 2, 3]))

# 37 open → abre un archivo (ejemplo con context manager, ya usado arriba).
with open("archivo.txt", "a", encoding="utf-8") as f:
    f.write("Probando append\n")

# 38 range → genera una secuencia de números.
for i in range(5):
    print(i)

# 39 enumerate → obtiene índice y valor al recorrer.
for i, v in enumerate(["a", "b"]):
    print(i, v)

# 40 zip → combina elementos de varias listas.
a = [1, 2]
b = ["uno", "dos"]
print(list(zip(a, b)))

# 41 map → aplica función a cada elemento.
nums = [1, 2, 3]
print(list(map(lambda x: x*2, nums)))

# 42 filter → filtra elementos según condición.
nums = [1, 2, 3, 4]
print(list(filter(lambda x: x%2==0, nums)))

# 43 sorted → devuelve lista ordenada.
print(sorted([3, 1, 2]))

# 44 reversed → invierte una secuencia.
print(list(reversed([1, 2, 3])))

# 45 super → accede a métodos de la clase padre.
class A:
    def saludo(self):
        print("Hola desde A")
class B(A):
    def saludo(self):
        super().saludo()
        print("Hola desde B")
B().saludo()

# 46 __init__ → constructor de una clase.
class Persona2:
    def __init__(self, nombre):
        self.nombre = nombre
p2 = Persona2("Ana2")
print(p2.nombre)

# 47 dir → lista atributos y métodos disponibles.
print("dir lista:", dir([])[:10])  # muestro solo los primeros para resumir

# 48 help → mostramos la docstring en lugar de abrir el pager interactivo.
print("Docstring de print:", print.__doc__.splitlines()[0])

# 49 id → devuelve la dirección en memoria de un objeto.
x_val = 5
print("id:", id(x_val))

# 50 type → muestra el tipo de dato de un objeto.
print(type(3.14))

# 51 min → mínimo de una secuencia
print("min:", min([3, 1, 2]))

# 52 max → máximo de una secuencia
print("max:", max([3, 1, 2]))

# 53 sum → suma de elementos
print("sum:", sum([1, 2, 3]))

# 54 all → True si todos los elementos son verdaderos
print("all:", all([True, True, 1]))

# 55 any → True si algún elemento es verdadero
print("any:", any([False, True, 0]))

# 56 abs → valor absoluto
print("abs:", abs(-5))

# 57 pow → potencia (a**b)
print("pow:", pow(2, 3))

# 58 round → redondeo
print("round:", round(3.14159, 2))

# 59 format → formateo con especificadores
print("format:", format(3.14159, ".2f"))

# 60 f-strings → interpolación moderna
nombre = "Ana"
print(f"Hola {nombre} — ejemplo de f-string")

# 61 str → conversión a cadena
print("str:", str(123))

# 62 int → conversión a entero
print("int:", int("123"))

# 63 float → conversión a flotante
print("float:", float("3.5"))

# 64 complex → número complejo
cnum = complex(1, 2)
print("complex:", cnum)

# 65 bool → tipo booleano
print("bool:", bool(0), bool(1))

# 66 list → constructor de listas
print("list:", list((1, 2, 3)))

# 67 dict → constructor de diccionarios
print("dict:", dict(a=1, b=2))

# 68 set → conjunto
print("set:", set([1, 2, 2]))

# 69 tuple → tupla
print("tuple:", tuple([1, 2]))

# 70 frozenset → conjunto inmutable
print("frozenset:", frozenset([1, 2, 2]))

# 71 bytes → secuencia inmutable de bytes
print("bytes:", bytes("hola", "utf-8"))

# 72 bytearray → secuencia mutable de bytes
ba = bytearray(b"abc")
ba.append(100)
print("bytearray:", ba)

# 73 slice → crear rebanadas
s = slice(1, 4)
print("slice:", [0,1,2,3,4][s])

# 74 getattr → obtener atributo con valor por defecto
class Ej:
    def __init__(self):
        self.x = 7
e = Ej()
print("getattr:", getattr(e, "x", None))

# 75 setattr → asignar atributo dinámicamente
setattr(e, "y", 20)
print("setattr:", e.y)

# 76 hasattr → comprobar existencia de atributo
print("hasattr:", hasattr(e, "y"))

# 77 delattr → eliminar atributo existente (seguro porque existe)
delattr(e, "y")
print("delattr: eliminado, hasattr y ->", hasattr(e, "y"))

# 78 callable → saber si un objeto es llamable
def ejemplo():
    return "soy llamable"
print("callable:", callable(ejemplo))

# 79 exec → ejecutar código en tiempo de ejecución (ejecución segura de ejemplo)
scope = {}
exec("z = 5", scope)
print("exec -> z:", scope.get("z"))

# 80 iter → obtener un iterador
it = iter([1, 2, 3])
print("iter + next:", next(it))

# 81 next → avanzar iterador (ya usado arriba y es seguro)
it2 = iter([9, 8])
print("next:", next(it2))

# 82 list comprehension → forma concisa de listas
lc = [x*2 for x in range(3)]
print("list comp:", lc)

# 83 dict comprehension → diccionario por comprensión
dc = {i: i*i for i in range(3)}
print("dict comp:", dc)

# 84 set comprehension → conjunto por comprensión
sc = {i for i in range(3)}
print("set comp:", sc)

# 85 generator expression → generador perezoso
gen = (x for x in range(3))
print("gen expr next:", next(gen))

# 86 itertools.chain → encadenar iterables
import itertools
print("chain:", list(itertools.chain([1], [2])))

# 87 collections.namedtuple → tupla con nombre
from collections import namedtuple
Point = namedtuple("Point", "x y")
p = Point(1, 2)
print("namedtuple:", p, p.x, p.y)

# 88 collections.deque → cola rápida
from collections import deque
dq = deque([1, 2])
dq.append(3)
print("deque:", list(dq))

# 89 collections.defaultdict → diccionario con valor por defecto
from collections import defaultdict
dd = defaultdict(int)
dd["a"] += 1
print("defaultdict:", dict(dd))

# 90 collections.OrderedDict → diccionario ordenado (útil conceptualmente)
from collections import OrderedDict
od = OrderedDict()
od["a"] = 1
od["b"] = 2
print("OrderedDict:", list(od.items()))

# 91 json → serializar / deserializar
import json
s = json.dumps({"a": 1})
print("json dumps:", s)
print("json loads:", json.loads(s))

# 92 re → expresiones regulares básicas
import re
m = re.search("a", "abc")
print("re search:", bool(m))

# 93 pathlib → manipulación de rutas moderna
from pathlib import Path
p = Path(".")
print("path exists:", p.exists())

# 94 os → operaciones de sistema (seguras)
import os
print("cwd:", os.getcwd())

# 95 os.path → combinar rutas
print("join:", os.path.join("foo", "bar"))

# 96 shutil → utilidades de alto nivel (ejemplo seguro)
import shutil
print("shutil which (si está en PATH):", shutil.which("python") is not None)

# 97 tempfile → crear directorio temporal (seguro y autolimpiante)
import tempfile
with tempfile.TemporaryDirectory() as tmpdir:
    print("tempdir creado:", tmpdir)

# 98 itertools.product → producto cartesiano (pequeño ejemplo)
print("product:", list(itertools.product([1, 2], ["a"])))

# 99 itertools.permutations → permutaciones pequeñas
print("permutations:", list(itertools.permutations([1, 2])))

# 100 vars → obtener dict de atributos de un objeto
class Cvars:
    def __init__(self):
        self.a = 1
cv = Cvars()
print("vars:", vars(cv))

# 101 globals → diccionario global (mostrar clave típica)
g = globals()
print("globals contiene 'print'?:", 'print' in g)

# 102 locals → diccionario local (tipo)
print("locals es dict?:", isinstance(locals(), dict))

# 103 memoryview → vista de bytes (ejemplo simple)
mv = memoryview(b"abc")
print("memoryview[0]:", mv[0])

# 104 callable (ya mostrado) → ejemplo con clase que implementa __call__
class CallMe:
    def __call__(self):
        return "called"
cm = CallMe()
print("CallMe es callable:", callable(cm), cm())

# 105 property → crear propiedad en clase
class Prop:
    def __init__(self, x):
        self._x = x
    @property
    def x(self):
        return self._x
pr = Prop(7)
print("property:", pr.x)

# 106 staticmethod → ejemplo
class EjStatic:
    @staticmethod
    def sal():
        return "hola"
print("staticmethod:", EjStatic.sal())

# 107 classmethod → ejemplo
class EjClass:
    value = 10
    @classmethod
    def get_value(cls):
        return cls.value
print("classmethod:", EjClass.get_value())

# 108 decorator → función que envuelve otra
def mi_deco(fn):
    def wrapper(*args, **kwargs):
        return "decorated -> " + str(fn(*args, **kwargs))
    return wrapper

@mi_deco
def saludo_simple():
    return "hola"
print("decorator:", saludo_simple())

# 109 contextlib.contextmanager → crear un context manager simple
from contextlib import contextmanager
@contextmanager
def mi_cm():
    print("enter")
    try:
        yield "recurso"
    finally:
        print("exit")

with mi_cm() as r:
    print("usando:", r)

# 110 csv → escritura/lectura en memoria con io.StringIO para evitar archivos
import csv, io
sio = io.StringIO()
writer = csv.writer(sio)
writer.writerow(["a", "b", "c"])
sio.seek(0)
reader = csv.reader(sio)
print("csv row:", next(reader))

# 111 argparse → definir parser (no parsear argumentos reales aquí)
import argparse
parser = argparse.ArgumentParser(description="Ejemplo seguro")
parser.add_argument("--num", type=int, default=0)
# No llamamos parser.parse_args() para no depender de sys.argv

# 112 sys → información del intérprete
import sys
print("python version:", sys.version.split()[0])

# 113 traceback → mostrar cómo obtener una cadena de excepción (sin lanzar)
import traceback
try:
    x = 1 + 1  # no error, pero podemos obtener formato de stack si fuera necesario
    tb = "".join(traceback.format_stack(limit=1))
    print("traceback snippet:", tb.strip().splitlines()[-1])
except Exception:
    pass

# 114 math funciones adicionales
print("math sin:", math.sin(0), "math log:", math.log(1))

# 115 random → números aleatorios (ejemplo seguro)
import random
print("random choice:", random.choice([1, 2, 3]))

# 116 time / datetime → fecha y hora
import datetime
print("today:", datetime.date.today())

# 117 enumerate (ya usado) → ejemplo con start
for idx, val in enumerate(["x", "y"], start=1):
    print("enumerate start=1:", idx, val)

# 118 ternary operator → expresion condicional
a = 5
print("ternary:", "par" if a % 2 == 0 else "impar")

# 119 slicing avanzado → paso en slice
print("slice paso:", list(range(10))[1:8:2])

# 120 documentation string / __doc__ → acceso a docstrings
def fdoc():
    """Función de ejemplo con docstring."""
    return None
print("docstring:", fdoc.__doc__.split()[0])
# 121 isinstance → verifica tipo de un objeto
print("isinstance:", isinstance(5, int))

# 122 issubclass → comprueba herencia
class A: pass
class B(A): pass
print("issubclass:", issubclass(B, A))

# 123 bin → representación binaria
print("bin:", bin(10))

# 124 oct → representación octal
print("oct:", oct(10))

# 125 hex → representación hexadecimal
print("hex:", hex(255))

# 126 ord → código unicode de un carácter
print("ord:", ord("A"))

# 127 chr → carácter a partir de código unicode
print("chr:", chr(65))

# 128 ascii → representación ASCII de un objeto
print("ascii:", ascii("á"))

# 129 repr → representación oficial de un objeto
print("repr:", repr("texto"))

# 130 hash → valor hash de un objeto
print("hash:", hash("clave"))

# 131 reversed iter → invertir iterador
print("reversed str:", "".join(reversed("hola")))

# 132 all + comprehensions → todos positivos
nums = [1, 2, 3]
print("all >0:", all(n > 0 for n in nums))

# 133 any + comprehensions → algún negativo
nums = [1, -2, 3]
print("any <0:", any(n < 0 for n in nums))

# 134 divmod → cociente y residuo
print("divmod:", divmod(7, 3))

# 135 round negativo → redondeo a decenas
print("round -1:", round(123, -1))

# 136 zip + dict → combinar en diccionario
claves = ["a", "b"]
vals = [1, 2]
print("zip dict:", dict(zip(claves, vals)))

# 137 enumerate con tupla → índice y valor
for i, v in enumerate(("x", "y")):
    print("enum tupla:", i, v)

# 138 eval → evaluar expresión segura
expr = "2+3"
print("eval:", eval(expr, {"__builtins__": None}, {}))

# 139 open read → leer archivo seguro
with open("archivo.txt", "r", encoding="utf-8") as f:
    print("contenido archivo:", f.readline().strip())

# 140 vars función → inspección atributos
def demo_func(): pass
print("vars func:", hasattr(demo_func, "__name__"))

# 141 setattr dinámico en clase
class D: pass
d = D()
setattr(d, "nuevo", 99)
print("dinámico:", d.nuevo)

# 142 getattr con fallback
print("getattr con default:", getattr(d, "otro", "no existe"))

# 143 bool en colecciones vacías
print("bool([]):", bool([]), "bool([1]):", bool([1]))

# 144 isinstance múltiples tipos
print("int o float:", isinstance(3.5, (int, float)))

# 145 frozenset operaciones
fs = frozenset([1,2,3])
print("frozenset union:", fs | {3,4})

# 146 map con str.upper
print("map upper:", list(map(str.upper, ["a","b"])))

# 147 filter con lambda negativo
nums = [-1,0,1]
print("filter negativos:", list(filter(lambda x:x<0, nums)))

# 148 sorted con key
palabras = ["zzz","a","mm"]
print("sorted len:", sorted(palabras, key=len))

# 149 slice objeto con step negativo
print("slice negativo:", [0,1,2,3,4][::-1])

# 150 help sobre tipo básico (línea docstring)
print("help int:", int.__doc__.split()[0])


```
