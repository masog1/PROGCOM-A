```python
import numpy as np
import matplotlib.pyplot as plt

# Tiempo en segundos
t = np.linspace(0, 2, 500)

# Señal de presión arterial simulada (latido cardiaco)
presion = 80 + 40 * np.sin(2 * np.pi * 1.2 * t) * np.exp(-t/3)

# --- Problema: calcular métricas ---
PAS = np.max(presion)   # presión sistólica
PAD = np.min(presion)   # presión diastólica
PP = PAS - PAD          # presión de pulso

print("Presión Sistólica (PAS):", round(PAS,1), "mmHg")
print("Presión Diastólica (PAD):", round(PAD,1), "mmHg")
print("Presión de Pulso (PP):", round(PP,1), "mmHg")

# --- Graficar con anotaciones ---
plt.figure(figsize=(8,4))
plt.plot(t, presion, color="red", linewidth=2, label="Onda de presión")
plt.scatter(t[np.argmax(presion)], PAS, color="blue", label="PAS (máx)")
plt.scatter(t[np.argmin(presion)], PAD, color="green", label="PAD (mín)")
plt.title("Simulación de onda de presión arterial")
plt.xlabel("Tiempo (s)")
plt.ylabel("Presión (mmHg)")
plt.legend()
plt.grid(True)
plt.show()
```
